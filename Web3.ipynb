### 9. Scrape the details most watched tv series of all time from imdb.com.

```python
import selenium
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.action_chains import ActionChains
import time
```

```python
driver=webdriver.Chrome(r"C:\Users\eshan\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
url="https://www.imdb.com/list/ls095964455/"
driver.get(url)
driver.maximize_window()
```

```python
Name=[]
Year_span=[]
Genre=[]
Run_time=[]
Ratings=[]
Votes=[]
```

```python
try:
    name=driver.find_elements_by_xpath("//h3[@class='lister-item-header']//a")
    for i in name:
        nam=i.text
        Name.append(nam)
except NoSuchElementException:
    Name.append('-')
    
try:
    year=driver.find_elements_by_xpath("//span[@class='lister-item-year text-muted unbold']")
    for i in year:
        yr=i.text
        Year_span.append(yr)
except NoSuchElementException:
    Year_span.append('-')
    
try:
    genre=driver.find_elements_by_xpath("//p[@class='text-muted text-small']//span[@class='genre']")
    for i in genre:
        gen=i.text
        Genre.append(gen)
except NoSuchElementException:
    Genre.append('-')
    
try:
    time=driver.find_elements_by_xpath("//p[@class='text-muted text-small']//span[@class='runtime']")
    for i in time:
        tim=i.text
        Run_time.append(tim)
except NoSuchElementException:
    Run_time.append('-')
    
try:    
    ratings=driver.find_elements_by_xpath("//div[@class='ipl-rating-star small']//span[2]")
    for i in ratings:
        rate=i.text
        Ratings.append(rate)
except NoSuchElementException:
    Ratings.append('-')
    
try:    
    votes=driver.find_elements_by_xpath("//p[@class='text-muted text-small']//span[@name='nv']")
    for i in votes:
        vot=i.text
        Votes.append(vot)
except NoSuchElementException:
    Votes.append('-')
    
```

```python
import pandas as pd
list_of_movies=pd.DataFrame({})
list_of_movies['Name']=Name
list_of_movies['Year_span']=Year_span
list_of_movies['Genre']=Genre
list_of_movies['Run_time']=Run_time
list_of_movies['Ratings']=Ratings
list_of_movies['Votes']=Votes
```

```python
list_of_movies
```

### 10. Details of Datasets from UCI machine learning repositories.


```python
import selenium
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.action_chains import ActionChains
import time
```

```python
driver=webdriver.Chrome(r"C:\Users\eshan\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
url="https://archive.ics.uci.edu/"
driver.get(url)
driver.maximize_window()
```

```python
show_all=driver.find_element_by_link_text("View ALL Data Sets")
driver.implicitly_wait(10)
ActionChains(driver).move_to_element(show_all).click(show_all).perform()
```

```python
Dataset_name=[]
Data_type=[]
Task=[]
Attribute_type=[]
No_of_instances=[]
No_of_attribute=[]
Year=[]
```

```python
name=driver.find_elements_by_xpath("//table[@cellpadding='3']//tr//table[@border='1']//tr//td//table//td//p//a")
for i in name:
    data=i.text
    Dataset_name.append(data)

task=driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[2]/p")
for i in task:
    dat=i.text
    Task.append(dat)

datatype=driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[3]/p")
for i in datatype:
    data=i.text
    Data_type.append(data)
    
attribute=driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[4]/p")
for i in attribute:
    attr=i.text
    Attribute_type.append(attr)
    
instances=driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[5]/p")
for i in instances:
    inst=i.text
    No_of_instances.append(inst)
    
num_attr=driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[6]/p")
for i in num_attr:
    natt=i.text
    No_of_attribute.append(natt)

year=driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[7]/p")
for i in year:
    yr=i.text
    Year.append(yr)
    
```

```python
Task=Task[1:]
Data_type=Data_type[1:]
Attribute_type=Attribute_type[1:]
No_of_instances=No_of_instances[1:]
No_of_attribute=No_of_attribute[1:]
Year=Year[1:]
```

```python
import pandas as pd
UCI=pd.DataFrame({})
UCI['Dataset_name']=Dataset_name
UCI['Task']=Task
UCI['Data_type']=Data_type
UCI['Attribute_type']=Attribute_type
UCI['No_of_instances']=No_of_instances
UCI['No_of_attribute']=No_of_attribute
UCI['Year']=Year
```
